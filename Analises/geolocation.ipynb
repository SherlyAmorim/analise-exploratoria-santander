```python
import pandas as pd
import numpy as np
```

Código gerado para ler o arquivo csv obtido do dataSet

```python
df = pd.read_csv('../data/olist_geolocation_dataset.csv')
df
```

Baseada no código postal brasileiro, onde os 5 primeiros dígitos representam respectivamente:
1) região, 2) sub-região, 3) setor, 4) subsetor, 5) divisor de subsetor; 

Podemos filtrar pela coluna 'geolocation_zip_code_prefix' e contabilizar através dos códigos postais a quantidade de clientes por região. 

```python
cities = df[df['geolocation_zip_code_prefix'].astype(str).str.startswith('1020')]
qtd_por_cities = cities.shape[0] 
resultado = cities[['geolocation_zip_code_prefix', 'geolocation_city']]

print(f'Temos {qtd_por_cities} clientes na região do código postal informado')
print('')
print(resultado)
```

```python
state_custumers = df['geolocation_state'].unique()
list(np.sort(state_custumers))
```

Métodos em pandas utilizados para obter os nomes das cidades (sem duplicidade) que os clientes residem e com alguns tratamentos para evitar duplicidade por motivo de diferença de digitação, como letras maiúsculas e minusculas, acentos, etc.

```python
cities_custumers = df['geolocation_city'].str.lower().str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').unique()
list(cities_custumers)

```

```python
num_cities_custumers = pd.Series(cities_custumers).nunique()
print(f"Temos clientes em aproximadamente {num_cities_custumers} cidades brasileiras!")
```
